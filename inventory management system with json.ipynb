```python
#Importing Libarires
import json
import time

#Initializing the variables
sale = ' '

#Loading the inventory Records from JSON to string
ch = 'Y'
fd = open('Record.json','r')
js = fd.read()
fd.close()

#Converting String Records to JSON Dictionaries
record = json.loads(js)


#Displaying the menu
print("--------------------MENU---------------------")
for key in record.keys():
    print(key, record[key]['Name'], record[key]['price'], record[key]['Qn'])
print("---------------------------------------------")
print('')


#Taking user input about the Purchasing and User deatails
ui_name =str(input("Enter the Name    :"))
ui_mail =str(input("Enter the Email   :"))
ui_ph =str(input("Enter the Phone num :"))
ui_pr = str(input("Enter product ID : "))
ui_qn = int(input("Enter Quantiry   : "))

print("---------------------------------------------")
print('')

#If we're having equl or more quantity then user wants
if (int(record[ui_pr]['Qn']) >= ui_qn):
    
    #Displaying the Purshing Details
    print("Name      : ", record[ui_pr]["Name"])
    print("Price (Rs): ", record[ui_pr]["price"])
    print("Quantity  : ", ui_qn)
    print("---------------------------------------------")
    if (ui_qn * int(record[ui_pr]["price"]) >= 5000):
        discount = (ui_qn * int(record[ui_pr]["price"]) * (5/100))
        print("Billing   : ", ui_qn * int(record[ui_pr]["price"]) - discount, "Rs")                      
    else:
        print("Billing   : ", ui_qn * int(record[ui_pr]["price"]), "Rs")
    print("---------------------------------------------")
        
    
    #Generating Sales Structure in CSV
    sale = ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]["Name"]+","+str(ui_qn)+","+str(record[ui_pr]["price"])+","+str(ui_qn * record[ui_pr]["price"])+","+time.ctime()+'\n'
    
    #Updating Records Dictionary
    record[ui_pr]['Qn'] = int(record[ui_pr]['Qn']) - ui_qn

#If we're not having quantity then user wants
else:
    
    #Displaying Info about the Quantity
    print("Sorry, We're not having enough quanity of product in our Inventory.")
    print("We're only having " + str(record[ui_pr]['Qn']) + " quantity.")
    print("---------------------------------------------")
    
    #If user wants to purchase whole Quantity for that product
    ch == str(input("Press Y to purchase: "))
    
    if(ch == "Y" or ch == 'y'):
    
        #Displaying the Purshing Details
        
        print("---------------------------------------------")
        print("Name      : ", record[ui_pr]["Name"])
        print("Price (Rs): ", record[ui_pr]["price"])
        print("Quantity  : ", record[ui_pr]['Qn'])
        print("---------------------------------------------")
        if (ui_qn * int(record[ui_pr]["price"]) >= 5000):
            discount = (ui_qn * int(record[ui_pr]["price"]) * (5/100))
            print("Billing   : ", ui_qn * int(record[ui_pr]["price"]) - discount, "Rs")                      
        else:
            print("Billing   : ", ui_qn * int(record[ui_pr]["price"]), "Rs")
        print("---------------------------------------------")

        #Genarting Sales Structure in CSV
        sale = '1'+","+ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]["Name"]+","+str(record[ui_pr]['Qn'])+","+str(record[ui_pr]["price"])+","+str(record[ui_pr]['Qn'] * record[ui_pr]["price"])+","+time.ctime()+'\n'
        
        #Updating Records Dictionary
        record[ui_pr]['Qn'] = 0
        
        fd = open('sales.txt','a')
        fd.write(sale)
        fd.close()
        
    else:
        print("Thanks!")
        
#Converting Dictionary to String
js = json.dumps(record)

#saving Records in Json File
fd = open('Record.json','w')
fd.write(js)
fd.close()

#Saving Records in Sales File
fd = open('sales.txt','a')
fd.write(sale)
fd.close()

print('')
print("---------------------------------------------")
print("  Thanks for your order, Inventory Updated!  ")
print("---------------------------------------------")
```
